## Introduction

In this notebook I'll show how to assign new value for each unique group in Pandas DataFrame.

````python
#collapse-show
import pandas as pd
````

````python
data = {"ID": [1, 1, 1, 2, 2, 3, 3], "class": ["Lower", "Moderate", "Moderate", "Lower", "High", "High", "Lower"]}
````

Let convert this dictionary into Pandas DataFrame and see what data we have.

````python
df = pd.DataFrame(data)
````

````python
df
````

Now, let assign new values according to ```class``` column and then rearrange ```class``` column in order to assign new group names.

````python
value_map = {"Lower": 33, "Moderate": 34, "High": 40}

df["new"] = df["class"].map(value_map)

df
````

````python
df["Max"] = df.set_index("class").groupby("ID")["new"].transform("idxmax").values

df
````

If we want only first values per group

````python
df["Max"] = df["Max"].mask(df["ID"].duplicated(), "")

df
````
